```python
import matplotlib as plt
%matplotlib inline
import pandas as pd
```

```python
data=pd.read_csv('Cairo.csv')
data.head()
```

```python
fig = plt.pyplot.figure()

ax=fig.add_axes([0,0,1,1])

plt.pyplot.xticks(rotation=45)
fig.suptitle('Comparison between Average Global Temprature and Cairo City average Global temprature',y=-0.25)
ax.plot(data['Years'].dropna(),data['Cairo Temprature'].dropna(),'r', label="Cairo Temprature")
plt.pyplot.xlabel="Years"
ax.plot(data['Years'].dropna(), data['Global Temprature'].dropna(),'b', label="Global Temprature")
ax.set_xlabel('Years')
ax.set_ylabel('Average Temprature(celsuis)')
ax.legend()
plt.pyplot.show()
#plt.pyplot.subplot(data['Years'],data['Global Temprature'],'r')
```
